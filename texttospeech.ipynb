```python
from gtts import gTTS
import IPython.display as ipd
```

```python
def text_to_speech(text):
    tts=gTTS(text=text, lang='en')
    audio_file="output.mp3"
    tts.save(audio_file)
    return ipd.Audio(audio_file)
```

```python
text='''Johny Johny, yes papa?
Eating sugar? No soni beta
Telling lies? No soni beta open your muh
Open your mouth
Ha, ha, ha
ha, ha, ha'''
text_to_speech(text)
```
